# (G)LMMs, GLMs, and ANOVA

```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches
```

```python
verts = [(0.0, 0.0), (0.50, 2.0), (2.0, 2.0), (1.5, 0), (0.0, 0.0)]

codes = [Path.MOVETO, Path.LINETO, Path.LINETO, Path.LINETO, Path.CLOSEPOLY]

path = Path(verts, codes)

fig = plt.figure(figsize = (16, 8))

ax = fig.add_subplot(111)

patch = patches.PathPatch(path, facecolor = '#E69F00', lw = 0)

ax.add_patch(patch)

ax.set_title('(Generalized) Linear Mixed Model: parallelogram', fontsize = 21, fontweight = 'bold')

ax.axis('off')
ax.set_xlim(-1, 3);
ax.set_ylim(-1, 3);
```

# (Generalized) Linear Mixed Models
- general case  
- can be implemented, but can be difficult in the real world  
- at times poorly understood/under-specified  
- if you can implement in a defensible manner, great

# (Generalized) Linear Mixed Models
- rule of thumb for fixed and random effects  
    + fixed: affects mean  
    + random: affects variance  
    + real world: predictor dies both  
- balanced vs. unbalanced (missing observations vs. missing participation)  
- `Subject/Participant` is a random effect at the very least  
- discuss other experimental paradigms as they come up  

```python
verts = [(0.0, 0.0), (0.0, 2.0), (2.0, 2.0), (2.0, 0), (0.0, 0.0)]

codes = [Path.MOVETO, Path.LINETO, Path.LINETO, Path.LINETO, Path.CLOSEPOLY]

path = Path(verts, codes)

fig = plt.figure(figsize = (16, 8))

ax = fig.add_subplot(111)

patch = patches.PathPatch(path, facecolor = '#E69F00', lw = 0)

ax.add_patch(patch)

ax.set_title('Generalized Linear Model: rectangle', fontsize = 21, fontweight = 'bold')

ax.axis('off')
ax.set_xlim(-1, 3);
ax.set_ylim(-1, 3);
```

# Generalized Linear Models
- special case of the general  
- applies to many situations (if conditions are met)  
- some difficulties in implementing (e.g., overdispersion)  
- useful in many cases

```python
verts = [(0.0, 0.0), (0.0, 2.0), (1.0, 2.0), (1.0, 0.0), (0.0, 0.0)]

codes = [Path.MOVETO, Path.LINETO, Path.LINETO, Path.LINETO, Path.CLOSEPOLY]

path = Path(verts, codes)

fig = plt.figure(figsize = (16, 8))

ax = fig.add_subplot(111)

patch = patches.PathPatch(path, facecolor = '#E69F00', lw = 0)

ax.add_patch(patch)

ax.set_title('ANOVA: square', fontsize = 21, fontweight = 'bold')

ax.axis('off')
ax.set_xlim(-2, 3);
ax.set_ylim(-2, 3);
```

# Formula notation
- form: `response ~ predictor1 + predctor2 + ....`  
- symbols:  
    + `+`: include in model  
    + `-`: delete from model  
    + `*`: include predictors and calculate interactions  
    + `:`: interaction  
    + `^`: crossing to *nth* degree  
    + `|`: conditional on (goes in parentheses in R usually)  

For a clear explanation, [see this PDF](http://faculty.chicagobooth.edu/richard.hahn/teaching/FormulaNotation.pdf)
